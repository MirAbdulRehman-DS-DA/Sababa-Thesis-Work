```python
# Importing Libraries

from rdkit import Chem
from rdkit.Chem import PandasTools
import pandas as pd
import os
```

```python
# Load the SDF file
Atorvastatin_sdf_file = 'data\Group A\Atorvastatin\SDF\Atorvastatin-DB01076.sdf' # initializing Variable to store location for Atorvastatin
Metformin_sdf_file = 'data\Group A\Metformin\SDF\Metformin-DB00331.sdf' # initializing Variable to store location for Metformin
Omeprazole_sdf_file = 'data\Group A\Omeprazole\SDF\Omeprazole-DB00338.sdf' # initializing Variable to store location for Omeprazole

suppl = Chem.SDMolSupplier(Atorvastatin_sdf_file)
suppl = Chem.SDMolSupplier(Metformin_sdf_file)
suppl = Chem.SDMolSupplier(Omeprazole_sdf_file)

# Convert to Pandas DataFrame
Atorvastatin_df = PandasTools.LoadSDF(Atorvastatin_sdf_file)
Metformin_df = PandasTools.LoadSDF(Metformin_sdf_file)
Omeprazole_df = PandasTools.LoadSDF(Omeprazole_sdf_file)

# Display the DataFrame
# print(Atorvastatin_df.head())
```

```python
# Merge DataFrames
merged_df = pd.concat([Atorvastatin_df, Metformin_df, Omeprazole_df], ignore_index=True)

merged_df.head()
```

```python
merged_df.to_csv('merged.csv',index=False)
```
